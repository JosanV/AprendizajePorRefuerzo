```python
#imports
import numpy as np
import random
import pandas
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import matplotlib.pyplot as plt
from time import time
```

```python
def generar_r(matriz, casilla_final):
    n, m = matriz.shape #n filas y m columnas
    r = np.full((n*m, n*m), -1) #matriz R inicial de n^2 filas x m^2 columnas con valores a -1

    for i in range(len(matriz)):
        for j in range(len(matriz[i])):

            #Arriba izquierda
            try:
                if (i-1 < 0) or (j-1 < 0): raise IndexError("No menores a 0")      

                arrib_izq = matriz[i-1][j-1]
                if arrib_izq == casilla_final:
                    r[matriz[i][j]][arrib_izq] = 100
                else:
                    r[matriz[i][j]][arrib_izq] = 0

            except IndexError:
                pass 


            #Arriba centro
            try:
                if (i-1 < 0) or (j < 0): raise IndexError("No menores a 0")      

                arrib_cent = matriz[i-1][j]
                if arrib_cent == casilla_final:
                    r[matriz[i][j]][arrib_cent] = 100
                else:
                    r[matriz[i][j]][arrib_cent] = 0

            except IndexError:
                pass 


            #Arriba derecha
            try:
                if (i-1 < 0) or (j+1 < 0): raise IndexError("No menores a 0")      

                arrib_der = matriz[i-1][j+1]
                if arrib_der == casilla_final:
                    r[matriz[i][j]][arrib_der] = 100
                else:
                    r[matriz[i][j]][arrib_der] = 0

            except IndexError:
                pass 


            #Centro derecha
            try:
                if (i < 0) or (j+1 < 0): raise IndexError("No menores a 0")      

                cent_der = matriz[i][j+1]
                if cent_der == casilla_final:
                    r[matriz[i][j]][cent_der] = 100
                else:
                    r[matriz[i][j]][cent_der] = 0

            except IndexError:
                pass 


            #Abajo derecha
            try:
                if (i+1 < 0) or (j+1 < 0): raise IndexError("No menores a 0")

                abaj_der = matriz[i+1][j+1]
                if abaj_der == casilla_final:
                    r[matriz[i][j]][abaj_der] = 100
                else:
                    r[matriz[i][j]][abaj_der] = 0

            except IndexError:
                    pass

            #Abajo centro
            try:
                if (i+1 < 0) or (j < 0): raise IndexError("No menores a 0")

                abaj_cent = matriz[i+1][j]
                if  abaj_cent == casilla_final:
                    r[matriz[i][j]][ abaj_cent] = 100
                else:
                    r[matriz[i][j]][ abaj_cent] = 0

            except IndexError:
                pass


            #Abajo izquierda
            try:
                if (i+1 < 0) or (j-1 < 0): raise IndexError("No menores a 0")

                abaj_izq = matriz[i+1][j-1]
                if abaj_izq == casilla_final:
                    r[matriz[i][j]][abaj_izq] = 100
                else:
                    r[matriz[i][j]][abaj_izq] = 0

            except IndexError:
                pass


            #Centro izquierda
            try:
                if (i < 0) or (j-1 < 0): raise IndexError("No menores a 0")

                cent_izq = matriz[i][j-1]
                if cent_izq == casilla_final:
                    r[matriz[i][j]][cent_izq] = 100
                else:
                    r[matriz[i][j]][cent_izq] = 0

            except IndexError:
                pass
            
    return r
            
```

```python
def generar_q(matriz):
    n, m = matriz.shape #n filas y m columnas
    q = np.full((n*m, n*m), 0)
    return q
```

```python
def generar_solucion(casilla_inicial, casilla_final, q):
    casilla_actual = casilla_inicial
    pasos = [casilla_inicial]

    while(casilla_actual != casilla_final):
        fila = q[casilla_actual].tolist() #fila de las acciones en q de la casilla actual
        casilla_actual = fila.index(max(fila))
        pasos.append(casilla_actual)

    return pasos
```

```python
def q_learning_fase1(matriz, r, casilla_final, gamma, iteraciones):
    start_time = time()
    matriz = np.array(matriz) #convertir a matriz de numpy
    n, m = matriz.shape #a filas y b columnas
    a = n*m-1 #valor maximo del tablero
    
    #Matriz R y Q inicial
    #r = matriz_r
    q = generar_q(matriz)
    lista_rendimiento = []
    
    #Algoritmo Q Learning fase 1
    for it in range(0, iteraciones):
        #1 Obtener estado inicial aleatorio
        estado = random.randint(0,a) #estado random (inicial) de los valores de las casillas
        while(estado != casilla_final):
            #2 Obtener accion de todas las acciones posibles de ese estado
            acciones = ([i for i in range(0,len(r[estado])) if r[estado][i] >= 0]) #index acciones posibles del estado seleccionado
            accion = random.choice(acciones) #accion random de todas las posibles
            
            #3 Obtener acciones posibles de la accion seleccionada (nuevo estado)
            next_acciones = ([i for i in range(0,len(r[accion])) if r[accion][i] >= 0]) #acciones posibles de accion seleccionada
            
            #4 Se actualiza el valor de Q
            q[estado][accion] = r[estado][accion] + gamma*max(q[accion][i] for i in next_acciones)

            #5 La accion se convierte en el nuevo estado
            estado = accion
        
        
    #RENDIMIENTO
        acum = 0
        maximo_valor = -1
        for i in range(0,a):
            for j in range(0,a):
                acum = q[i][j] + acum
                if(q[i][j] >= maximo_valor):
                    maximo_valor = q[i][j]
        rendimiento = (acum/(maximo_valor+0.0000000000001))*100 #Se le suma ese valor por si el maximo es 0, que no divida entre 0
        lista_rendimiento.append(rendimiento)
    lista_indices = ([i for i in range(0,len(lista_rendimiento))])
    print("Rendimiento")
    plt.plot(lista_indices, lista_rendimiento)
    plt.show()
    elapsed_time = time() - start_time
    print("Tiempo en segundos: " + str(elapsed_time))
    return q
```

```python
def q_learning_fase2(matriz, r, casilla_final, gamma, epsilon_inicial, alpha, iteraciones):
    start_time = time()
    matriz = np.array(matriz) #convertir a matriz de numpy
    n, m = matriz.shape #a filas y b columnas
    a = n*m-1 #valor maximo del tablero
    
    #Matriz R y Q inicial
    #r = generar_r(matriz, casilla_final)
    q = generar_q(matriz)
    lista_rendimiento = []
    
    
    #Algoritmo Q Learning fase 2
    epsilon = epsilon_inicial
    for it in range(0, iteraciones):
        #1 Obtener estado inicial aleatorio
        estado = random.randint(0,a) #estado random (inicial) de los valores de las casillas
        
        
        
        while(estado != casilla_final):
            #2 Obtener accion de todas las acciones posibles de ese estado
            acciones = ([i for i in range(0,len(r[estado])) if r[estado][i] >= 0]) #index acciones posibles del estado seleccionado
            accion = 0 #Inicializar variable
            random_num = random.uniform(0,1) #numero aleatorio de 0 a 1
            if(random_num < epsilon):
                accion = random.choice(acciones) #accion random de todas las posibles
            else:
                maximo = -1 # -1 para que lo sustituya por cualquier numero que encuentre primero
                index_maximo = []
                for j in acciones:
                    if(q[estado][j] > maximo):
                        index_maximo.clear
                        index_maximo.append(j)
                        maximo = q[estado][j]
                    elif(q[estado][j] == maximo):
                        index_maximo.append(j)
                accion = random.choice(index_maximo)
                        
                        
            #3 Obtener valores en Q de las acciones posibles de la accion seleccionada (nuevo estado)
            next_acciones = ([i for i in range(0,len(r[accion])) if r[accion][i] >= 0]) #acciones posibles de accion seleccionada
            lista_q = []
            for i in next_acciones:
                lista_q.append(q[accion][i]) #Valores de esas acciones en Q
            
            #4 Se actualiza el valor de Q
            q[estado][accion] = r[estado][accion] + gamma*max(lista_q)

            #5 La accion se convierte en el nuevo estado
            estado = accion
        epsilon = epsilon * alpha
        
    #RENDIMIENTO
        acum = 0
        maximo_valor = -1
        for i in range(0,a):
            for j in range(0,a):
                acum = q[i][j] + acum
                if(q[i][j] >= maximo_valor):
                    maximo_valor = q[i][j]
        rendimiento = (acum/(maximo_valor+0.0000000000001))*100 #Se le suma ese valor por si el maximo es 0, que no divida entre 0
        lista_rendimiento.append(rendimiento)
    lista_indices = ([i for i in range(0,len(lista_rendimiento))])
    print("Rendimiento")
    plt.plot(lista_indices, lista_rendimiento)
    plt.show()
    elapsed_time = time() - start_time
    print("Tiempo en segundos: " + str(elapsed_time))
    return q
```

```python
def generar_matriz_aleatoria(n, m):
    #n filas, m columnas
    k = n*m
    elements = random.sample(range(k), k)
    r = [elements[index : index + m] for index in (m * i for i in range(n))]
    r = np.array(r)
    return r
```

```python
def q_learning_fase1_3(matriz, r, casilla_final, gamma, iteraciones):
    start_time = time()
    matriz = np.array(matriz) #convertir a matriz de numpy
    n, m = matriz.shape #a filas y b columnas
    a = n*m-1 #valor maximo del tablero
    
    #Matriz R y Q inicial
    #r = matriz_r
    q = generar_q(matriz)
    lista_rendimiento = []
    
    #Algoritmo Q Learning fase 1
    for it in range(0, iteraciones):
        #1 Obtener estado inicial aleatorio
        estado = random.randint(0,a) #estado random (inicial) de los valores de las casillas
        while(estado != casilla_final):
            #2 Obtener accion de todas las acciones posibles de ese estado
            acciones = ([i for i in range(0,len(r[estado])) if r[estado][i] >= 0]) #index acciones posibles del estado seleccionado
            accion = random.choice(acciones) #accion random de todas las posibles
            
             #3 Obtener acciones posibles de la accion seleccionada (nuevo estado)
            next_acciones = ([i for i in range(0,len(r[accion])) if r[accion][i] >= 0]) #acciones posibles de accion seleccionada
            
            #4 Se actualiza el valor de Q
            q[estado][accion] = r[estado][accion] + gamma*max(q[accion][i] for i in next_acciones)


            #5 La accion se convierte en el nuevo estado
            estado = accion
            
    
     
    #RENDIMIENTO
        acum = 0
        maximo_valor = -1
        for i in range(0,a):
            for j in range(0,a):
                acum = q[i][j] + acum
                if(q[i][j] >= maximo_valor):
                    maximo_valor = q[i][j]
        rendimiento = (acum/(maximo_valor+0.0000000000001))*100 #Se le suma ese valor por si el maximo es 0, que no divida entre 0
        lista_rendimiento.append(rendimiento)
    lista_indices = ([i for i in range(0,len(lista_rendimiento))])
    print("Rendimiento")
    plt.plot(lista_indices, lista_rendimiento)
    plt.show()
    elapsed_time = time() - start_time
    print("Tiempo en segundos: " + str(elapsed_time))
    return q
```

```python
def q_learning_fase2_3(matriz, r, casilla_final, gamma, epsilon_inicial, alpha, iteraciones):
    start_time = time()
    matriz = np.array(matriz) #convertir a matriz de numpy
    n, m = matriz.shape #a filas y b columnas
    a = n*m-1 #valor maximo del tablero
    
    #Matriz R y Q inicial
    #r = generar_r(matriz, casilla_final)
    q = generar_q(matriz)
    lista_rendimiento = []
    
    #Algoritmo Q Learning fase 2
    epsilon = epsilon_inicial
    for it in range(0, iteraciones):
        #1 Obtener estado inicial aleatorio
        estado = random.randint(0,a) #estado random (inicial) de los valores de las casillas
        
        
        
        while(estado != casilla_final):
            #2 Obtener accion de todas las acciones posibles de ese estado
            acciones = ([i for i in range(0,len(r[estado])) if r[estado][i] >= 0]) #index acciones posibles del estado seleccionado
            accion = 0
            random_num = random.uniform(0,1) #numero aleatorio de 0 a 1
            if(random_num < epsilon):
                accion = random.choice(acciones) #accion random de todas las posibles
            else:
                maximo = -1 # -1 para que lo sustituya por cualquier numero que encuentre primero
                index_maximo = []
                for j in acciones:
                    if(q[estado][j] > maximo):
                        index_maximo.clear
                        index_maximo.append(j)
                        maximo = q[estado][j]
                    elif(q[estado][j] == maximo):
                        index_maximo.append(j)
                accion = random.choice(index_maximo)
                        
                        
            #3 Obtener valores en Q de las acciones posibles de la accion seleccionada (nuevo estado)
            next_acciones = ([i for i in range(0,len(r[accion])) if r[accion][i] >= 0]) #acciones posibles de accion seleccionada
            lista_q = []
            for i in next_acciones:
                lista_q.append(q[accion][i]) #Valores de esas acciones en Q
            
            #4 Se actualiza el valor de Q
            q[estado][accion] = r[estado][accion] + gamma*max(lista_q)

            #5 La accion se convierte en el nuevo estado
            estado = accion
        epsilon = epsilon * alpha
        
    #RENDIMIENTO
        acum = 0
        maximo_valor = -1
        for i2 in range(0,a):
            for j2 in range(0,a):
                acum = q[i2][j2] + acum
                if(q[i2][j2] >= maximo_valor):
                    maximo_valor = q[i2][j2]
        rendimiento = (acum/(maximo_valor+0.0000000000001))*100 #Se le suma ese valor por si el maximo es 0, que no divida entre 0
        lista_rendimiento.append(rendimiento)
    lista_indices = ([i for i in range(0,len(lista_rendimiento))])
    print("Rendimiento: ")
    plt.plot(lista_indices, lista_rendimiento)
    plt.show()
    #Tiempo ejecucion (depende de cada maquina)
    elapsed_time = time() - start_time
    print("Tiempo en segundos: " + str(elapsed_time))
        
    return q
```
